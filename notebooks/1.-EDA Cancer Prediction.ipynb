### **Train Test Split**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=42)
```

### **Preprocessing Pipeline**

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
numerical_pipeline= Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler())
    ]
)
```

```python
X_train= numerical_pipeline.fit_transform(X_train)
X_test= numerical_pipeline.transform(X_test)
```

```python
columns= numerical_pipeline.get_feature_names_out()
columns
```

```python
X_train= pd.DataFrame(X_train, columns=columns)
X_test= pd.DataFrame(X_test, columns=columns)
```

```python
X_train.head()
```

```python
X_test.head()
```

### **Model Training**

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
models= {
    'LogisticRegression': LogisticRegression(),
    'DecisionTreeClassifier': DecisionTreeClassifier(),
    'RandomForestClassifier': RandomForestClassifier(),
    'GradientBoostingClassifier': RandomForestClassifier(),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'SVM': SVC()
}
```

```python
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score

# Lists are Created to Store the Training F1 scores, 
# Training F1 Cross-Validated Scores, and Testing F1 Scores, Respectively.
tr= []
tr_cv= []
ts= []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred_tr= model.predict(X_train)
    y_pred_ts= model.predict(X_train)
    tr_f1= f1_score(y_train, y_pred_tr)
    ts_f1= f1_score(y_train, y_pred_ts)
    scores= cross_val_score(model, X_train, y_train, cv=5, scoring='f1')
    f1_cv= scores.mean()
    tr.append(tr_f1)
    ts.append(ts_f1)
    tr_cv.append(f1_cv)
    print(f'Model Name : {name}\n')
    print(f'Training F1 Score : {tr_f1:.4f}')
    print(f'Testing F1 Score  : {ts_f1:.4f}')
    print(f'Training F1 Cross Validated : {f1_cv:.4f}')
    print('\n============================================\n')
```

### **Creating DataFrame for Algorithms Scores**

```python
algo_eval= {
    'model':list(models.keys()),
    'Training_F1':tr,
    'Testing_F1':ts,
    'Cross_Val':tr_cv
}
```

```python
df_eval = pd.DataFrame(algo_eval)
df_eval
```

```python
df_eval.sort_values(by='Cross_Val',ascending=False)
```

### **Select Best Model Logistic Regression**

```python
params = {'C': [0.1, 1, 10, 100, 1000]}
```

```python
from sklearn.model_selection import GridSearchCV
gscv= GridSearchCV(LogisticRegression(), param_grid=params, cv=5, scoring='f1')
gscv.fit(X_train, y_train)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_lr = gscv.best_estimator_
best_lr
```

### **Evaluating the Model**

```python
y_pred_tr = best_lr.predict(X_train)
y_pred_ts = best_lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cf= confusion_matrix(y_test, y_pred_ts)
cfd= ConfusionMatrixDisplay(cf, display_labels=best_lr.classes_)
cfd.plot()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_ts))
```
